```python

from PreprocessAssistment import *

from IRT import IRT
from BPMFSkillEncoded import *

import matplotlib.pyplot as plt


def PlotLoss(model, model_name):
    plt.plot(range(model.maxepoch), model.logloss_train, marker='o', label='Training Data')
    plt.plot(range(model.maxepoch), model.logloss_test, marker='v', label='Test Data')
    plt.plot(range(model.maxepoch), model.auc_train, marker='*', label='AUC: Train Data')
    plt.plot(range(model.maxepoch), model.auc_test, marker='*', label='AUC: Test Data')
    plt.title('The Truncated Assitment Dataset Learning Curve ' + model_name)
    plt.xlabel('Number of Epochs')
    plt.ylabel('RMSE')
    plt.legend()
    plt.grid()
    plt.show()
```

```python

file_path = "/home/lxu/Documents/StudentLearningProcess/skill_builder_data_corrected_withskills_section.csv"
data, num_skills, prob_skill_map = PreprocessAssistmentSkillBuilder(file_path)
print('ASSISTment 09-10 statistics ', data.shape[0],  len(np.unique(data['user_id'])), len(np.unique(data['problem_id'])), num_skills)
# prepare data
datanp = data.values
trainnp, testnp = train_test_split(datanp, test_size=0.2)
#
columnNames = list(data.head(0))
train = pd.DataFrame(data=trainnp, columns=columnNames)
test = pd.DataFrame(data=testnp, columns=columnNames)
```

```python

num_user = len(np.unique(data['user_id']))
num_prob = len(np.unique(data['problem_id']))
prob_cnt_per_user = np.zeros((num_user))
for i in range(data.shape[0]):
    prob_cnt_per_user[data.loc[i, 'user_id']] += 1

plt.hist(prob_cnt_per_user)
plt.show()
print('mean number of problems per user: ',np.mean(prob_cnt_per_user), 'median number of problems per user: ', np.median(prob_cnt_per_user))
```

```python
user_cnt_per_user = np.zeros((num_prob))
for i in range(data.shape[0]):
    user_cnt_per_user[data.loc[i, 'problem_id']] += 1
plt.hist(user_cnt_per_user)
plt.show()
print('mean number of problems per user: ',np.mean(user_cnt_per_user), 'median number of problems per user: ', np.median(user_cnt_per_user))
```

```python
# model: global + user + skill 
name = 'baseline + MF'
model = IRT()
model.set_params({ "epsilon": 1, "_lambda": 0.2, "momentum": 0.5, "maxepoch": 50, "num_batches": 300,
                            "batch_size": 1000, 'multi_skills': True, 'problem': False,})
model.fit(train, test, len(np.unique(data['user_id'])), num_skills, len(np.unique(data['problem_id']))) 

PlotLoss(model, name)
```
